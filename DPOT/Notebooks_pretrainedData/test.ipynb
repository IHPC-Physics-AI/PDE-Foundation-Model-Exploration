```python
# slice_extrap_seed.py
import h5py, numpy as np, os

src_dir = '/Volumes/T7/PDEBench/2D/diffusion-reaction/DPOT_Processed/test'
dst_dir = '/Volumes/T7/PDEBench/2D/diffusion-reaction/DPOT_Processed/test_extrap'
os.makedirs(dst_dir, exist_ok=True)

for i in range(20):
    with h5py.File(f'{src_dir}/data_{i}.hdf5', 'r') as f:
        data = f['data'][:]
    sliced = data[..., 90:101, :]
    with h5py.File(f'{dst_dir}/data_{i}.hdf5', 'w') as f:
        f.create_dataset('data', data=sliced)
    print(f'[{i}] {data.shape} -> {sliced.shape}')
```
